```python
import asyncio
import edge_tts
from IPython.display import display, Audio

#TEXT = "这是一个使用edge_tts进行文本到语音转换的简化示例。您可以使用音频控件来播放这段文字。"
#VOICE = "zh-CN-XiaoxiaoNeural"

TEXT = "The campaign said some of its internal communications had been hacked and suggested Iran was responsible and seeking to undermine the former president’s prospects in the November election."

VOICE = "en-US-AriaNeural"

async def text_to_speech_stream():
    communicate = edge_tts.Communicate(TEXT, VOICE)
    audio_data = b""
    async for chunk in communicate.stream():
        if chunk["type"] == "audio":
            audio_data += chunk["data"]
    return audio_data

async def main():
    audio_data = await text_to_speech_stream()
    
    # 显示音频控件
    display(Audio(audio_data, autoplay=False))

# 在Jupyter Notebook中运行
await main()
```
